## Theoretical maximum number of drugs

### Import packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from matplotlib.colors import ListedColormap
from core.core import *
```

### TMT-11

```python
n_pools = 9
n_replicates = 3
max_shared_pools, avg_shared_pools = [], []
y1, y2 = [], []
for n_drugs in tqdm(range(10, 21)):
    avg_max_shared_pools = []
    for k in range(10):
        pool_matrix = generate_sensing_matrix(n_pools = n_pools, n_drugs = n_drugs, n_replicates = n_replicates, plot_his = False)
        shared_pools = np.dot(pool_matrix.T, pool_matrix) - n_replicates * np.eye(n_drugs)
        shared_pools = np.max(shared_pools, axis = 0)
        max_shared_pools.append(np.max(shared_pools))
        avg_shared_pools.append(np.mean(shared_pools))
    y1.append(np.mean(avg_shared_pools))
    y2.append(np.mean(max_shared_pools))

x = np.arange(10, 21)
plt.figure(dpi = 300)
plt.plot(x, y1, marker = 'D', label = 'Avg shared pools')
plt.plot(x, y2, marker = 'D', label = 'Max shared pools')
plt.xlabel('Num of drugs')
plt.ylabel('Avg shared pools')
plt.legend()
```
